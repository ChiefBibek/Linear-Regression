Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df = pd.read_csv("canadaIncome.csv")
```

```python
df.head()
```

```python
plt.ylabel('Year')
plt.xlabel('per capita income (US$)')
plt.scatter(df.year,df["per capita income (US$)"],color='r')
```

```python
new_df = df.drop('per capita income (US$)',axis=1)
new_df
```

```python
income = df['per capita income (US$)']
income
```

```python
#crate a linear regression model
model = linear_model.LinearRegression()
model.fit(new_df,income)
```

```python
#predict
model.predict([[2024]])
```

```python
model.coef_
```

```python
model.intercept_
```

Y = m * X + b (m is coefficient and b is intercept)

```python
model.score(new_df,income)
```
